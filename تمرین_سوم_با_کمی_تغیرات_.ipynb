```python
import json
from sympy import sympify, Symbol

class Node:
    def __init__(self, degree):
        self.degree = degree
        self.coefficients = []
        self.next = None

class PolynomialList:
    def __init__(self):
        self.head = Node(None)
        self.head.next = self.head
        self.n = 0

    def get(self, degree):
        x = self.head.next
        while x != self.head and x.degree != degree:
            x = x.next
        if x == self.head:
            raise Exception('Polynomial with this degree not found')
        return x

    def insert_after(self, x, coefficients):
        y = Node(len(coefficients) - 1)
        y.coefficients = coefficients
        self.n += 1
        y.next = x.next
        x.next = y
        return y

    def delete(self, x):
        if self.is_empty():
            raise Exception('List is empty')
        self.n -= 1
        prev_node = self.head
        while prev_node.next != x:
            prev_node = prev_node.next
        prev_node.next = x.next
        return x

    def size(self):
        return self.n

    def is_empty(self):
        return self.n == 0

def get_coefficients(degree):
    x = Symbol('x')
    while True:
        try:
            expression = input(f"Enter the polynomial for degree {degree} (e.g., 'x**2 + 3*x + 2'): ")
            # Replace "^" with "**" for exponentiation
            expression = expression.replace("^", "**")
            # Use sympify to parse the expression
            polynomial = sympify(expression)
            # Check if the parsed expression is a polynomial
            if polynomial.is_polynomial(x):
                # Extract coefficients in the reversed order
                coefficients = [polynomial.coeff(x, i) for i in range(degree + 1)]
                return coefficients[::-1]
            else:
                raise ValueError
        except (ValueError, SyntaxError):
            print("Invalid input! Please enter a valid polynomial expression.")

def insert_polynomials(polynomial_list):
    print("Insert Polynomial Menu:")
    while True:
        parent_count = int(input("Enter the number of parent polynomials: "))
        for _ in range(parent_count):
            degree = int(input("Enter the degree of the polynomial: "))
            coefficients = get_coefficients(degree)
            parent_poly = polynomial_list.insert_after(polynomial_list.head, coefficients)

            child_count = int(input("Enter the number of child polynomials for this parent: "))
            for _ in range(child_count):
                coefficients = get_coefficients(degree)
                polynomial_list.insert_after(parent_poly, coefficients)

        if input("Do you want to continue adding polynomials? (yes/no): ").lower() != 'yes':
            break

def display_all_polynomials(polynomial_list):
    print("Polynomials in the list:")
    current_node = polynomial_list.head.next
    while current_node != polynomial_list.head:
        print(f"Degree: {current_node.degree}, Coefficients: {current_node.coefficients}")
        current_node = current_node.next

def multiply_polynomials(polynomial_list):
    print("Multiply Polynomials Menu:")
    parent_degree = int(input("Enter the degree of the parent polynomial to multiply: "))
    parent_node = polynomial_list.get(parent_degree)

    # Initialize the result polynomial coefficients with zeros
    result_degree = parent_degree * 2  # Maximum possible degree after multiplication
    result_polynomial = [0] * (result_degree + 1)

    current_node = parent_node.next
    while current_node != polynomial_list.head:
        child_polynomial = current_node.coefficients
        current_result_polynomial = [0] * (result_degree + 1)  # Temporary storage for the result

        # Perform polynomial multiplication
        for i in range(len(parent_node.coefficients)):
            for j in range(len(child_polynomial)):
                current_result_polynomial[i + j] += parent_node.coefficients[i] * child_polynomial[j]

        # Add the current result polynomial to the overall result
        for i in range(len(result_polynomial)):
            result_polynomial[i] += current_result_polynomial[i]

        current_node = current_node.next

    # Remove leading zeros from the result polynomial
    while result_polynomial[-1] == 0 and len(result_polynomial) > 1:
        result_polynomial.pop()

    # Update the polynomial list with the multiplied polynomial
    multiplied_poly = polynomial_list.insert_after(parent_node, result_polynomial)

    print(f"Result of multiplication for degree {parent_degree}:")
    print(f"Coefficients: {multiplied_poly.coefficients}")

    return multiplied_poly

def sum_polynomials(polynomial_list):
    print("Sum Polynomials Menu:")
    parent_degree = int(input("Enter the degree of the parent polynomial to sum: "))
    parent_node = polynomial_list.get(parent_degree)

    result_polynomial = [0] * (parent_degree + 1)

    current_node = parent_node.next
    while current_node != polynomial_list.head:
        child_polynomial = current_node.coefficients

        # Ensure both polynomials have the same degree
        if len(child_polynomial) != len(result_polynomial):
            print(f"Error: Polynomials must have the same degree for summation.")
            return

        # Perform polynomial summation
        for i in range(len(result_polynomial)):
            result_polynomial[i] += child_polynomial[i]

        current_node = current_node.next

    # Update the polynomial list with the summed polynomial
    summed_poly = polynomial_list.insert_after(parent_node, result_polynomial)

    print(f"Result of summation for degree {parent_degree}:")
    print(f"Coefficients: {summed_poly.coefficients}")

    return summed_poly


import sys

def exit_program(polynomial_list):
    print("Exiting...")
    sys.exit()


def main_menu():
    print("Main Menu:")
    print("1. Insert Polynomials")
    print("2. Multiply Polynomials")
    print("3. Sum Polynomials")
    print("4. Display Polynomials")
    print("5. Exit")

def main():
    polynomial_list = load_polynomials("polynomials.json")
    menu_options = {
        '1': insert_polynomials,
        '2': multiply_polynomials,
        '3': sum_polynomials,
        '4': display_all_polynomials,
        '5': exit_program
    }

    while True:
        main_menu()
        choice = input("Enter your choice: ")
        if choice in menu_options:
            if choice == '5':
                menu_options[choice](polynomial_list)
                break
            else:
                menu_options[choice](polynomial_list)
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
